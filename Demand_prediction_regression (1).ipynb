## Hyper parameter Tuning

```python
from sklearn.model_selection import cross_val_score
```

```python
NoofEstimator = [5, 10, 15, 20]
MinSampleLeaf = [1, 3, 5, 7]
MaxFeature = np.arange(0.1, 1.1, 0.1)
```

```python
best_score = []
for i in NoofEstimator:
    for j in MinSampleLeaf:
        for k in MaxFeature:
            result = [i, j, k]
            rfr = RandomForestRegressor(n_estimators = i, min_samples_leaf = j,
                                         max_features = k)
            rfr.fit(x_train, y_train)
            result.append(rfr.score(x_train, y_train))
            result.append(rfr.score(x_test, y_test))
            if len(best_score) == 0:
                best_score = result
            elif best_score[4] < result[4]:
                best_score = result
                print(best_score)
                
print('The final best result is:', best_score)
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
rf_rd = RandomizedSearchCV(estimator = rfr, param_distributions=dict(n_estimators = NoofEstimator,min_samples_leaf =MinSampleLeaf,max_features= MaxFeature),cv=5)

rf_rd.fit(x_train,y_train)
print(rf_rd.best_estimator_)
print(rf_rd.score(x_test,y_test))

imp_feature = rf_rd.best_estimator_.feature_importances_

print (imp_feature)


feature_list = list(x.columns)

feature_importance = sorted(zip(imp_feature,feature_list),reverse = True)
df5 = pd.DataFrame(feature_importance, columns = ['Importance','columns'])
```
